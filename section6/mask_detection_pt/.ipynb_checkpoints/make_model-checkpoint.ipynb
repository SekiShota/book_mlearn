### データセット読み込み

```python
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from torchvision.datasets import ImageFolder

# transformsを作成したのちテンソルに変換して読み込み
transform=transforms.Compose(transforms=[transforms.ToTensor()])

filepath="./image_train/"

dataset=ImageFolder(filepath, transform=transform)
print(len(dataset))
print(dataset.class_to_idx)
```

### バッチサイズを指定してデータローダ作成

```python
import matplotlib.pyplot as plt
import numpy as np

classes=np.array(["mask_off", "mask_on"])
n_img=16

# データローダ作成
dataloader=DataLoader(dataset, batch_size=22, shuffle=True)

dataiter=iter(dataloader)
img, label=next(dataiter)

plt.figure(figsize=(8,8))
for i in range(n_img):
    plt.subplot(4,4,i+1)
    plt.imshow(np.transpose(img[i],(1,2,0)))
    plt.title(classes[label[i]])
    plt.axis("off")
plt.show()
```

### データの拡張

```python
# アフィン変換
affine=transforms.RandomAffine(degrees=(-20,20), scale=(0.8,1.2))

# 左右反転
flip=transforms.RandomHorizontalFlip(1.0)

# 正規化
norm=transforms.Normalize(mean=[0.0,0.0,0.0], std=[1.0,1.0,1.0])


# 学習用のアフィン変換
affine_train=transforms.Compose([affine, norm])
# 学習用の左右反転
flip_train=transforms.Compose([flip, norm])
# 学習用のアフィン変換＋左右反転
arg_train=transforms.Compose([affine, flip, norm])

# 検証用の正規化
norm_valid=transforms.Compose([norm])
```

### データの水増し

```python
import torch
import random
from torch.utils.data import Subset, random_split

class MySubset(torch.utils.data.Dataset):
    def __init__(self, dataset, indices, transform=None):
        self.dataset = dataset
        self.indices = indices
        self.transform = transform

    def __getitem__(self, idx):
        img, label = self.dataset[self.indices[idx]]
        if self.transform:
            img = self.transform(img)

        return img, label

    def __len__(self):
        return len(self.indices)


# 学習データ、検証データに 8:2 の割合でデータセットを分割する。
train_size = int(0.8 * len(dataset))
valid_size=len(dataset)-train_size
# データの偏りをなくすための工夫としてインデックスをシャッフルする
indices=np.arange(len(dataset))
random.shuffle(indices)

#学習用データのオリジナル
train_dataset_original=Subset(dataset, indices[:train_size])
# 学習用データのアフィン変換
train_dataset_affine = MySubset(dataset, indices[:train_size], affine_train)
# 学習用データの左右反転
train_dataset_flip = MySubset(dataset, indices[:train_size], flip_train)
# 学習用データのアフィン変換+左右反転
train_dataset_arg = MySubset(dataset, indices[:train_size], arg_train)
# 学習用データ（オリジナル＋拡張）
train_dataset=train_dataset_original+train_dataset_affine+train_dataset_flip+train_dataset_arg

# 検証用データ
valid_dataset = MySubset(dataset, indices[train_size:], norm_valid)


print(f"full: {len(dataset)}, train: {len(train_dataset)}, valid: {len(valid_dataset)}")
```

```python
# データローダの作成
train_loader=DataLoader(train_dataset, batch_size=128, shuffle=True)
valid_loader=DataLoader(valid_dataset, batch_size=len(valid_dataset), shuffle=False)
```

```python
from torch.utils.data import DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np

dataiter = iter(train_loader)  # イテレータ
images, labels = next(dataiter)  # 最初のバッチを取り出す

plt.figure(figsize=(8,8))  # 画像の表示サイズ
for i in range(n_img):
    plt.subplot(4,4,i+1)
    plt.imshow(np.transpose(images[i], (1, 2, 0)))  # チャンネルを一番後ろに
    label=classes[labels[i]]
    plt.title(label)
    plt.tick_params(labelbottom=False, labelleft=False, bottom=False, left=False)  # ラベルとメモリを非表示に

print(images[15].size())
plt.show()
```

### アーキテクチャの定義

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=nn.Conv2d(3,16,5)
        self.pool=nn.MaxPool2d(2,2);
        self.conv2=nn.Conv2d(16,16,5)
        self.fc1=nn.Linear(19*19*16,256)
        self.dropout=nn.Dropout(0.5)
        self.fc2=nn.Linear(256,2)
        
    def forward(self, x):
        x=self.pool(F.relu(self.conv1(x)))
        print(x.size())
        x=self.pool(F.relu(self.conv2(x)))
        print(x.size())
        x=x.view(-1, 19*19*16)
        x=F.relu(self.fc1(x))
        x=self.dropout(x)
        x=self.fc2(x)
        return x
    
net=Net()
print(net)     
```

### モデルの学習

```python
from torch import optim

#交差エントロピー損失
loss_function=nn.CrossEntropyLoss()
#最適化アルゴリズム
optimizer=optim.Adam(net.parameters())
#損失のログ
record_loss_train=[]
record_loss_valid=[]

#学習
x_valid, t_valid=next(iter(valid_loader))
# x_valid, t_valid=x_valid.cuda(), t_valid.cuda()

for i in range(15):
  ###学習モード###
    net.train()
    loss_train=0
    for j, (x,t) in enumerate(train_loader):
        #順伝播
        y=net(x)
        #誤差の算出
        loss=loss_function(y, t)
        #誤差の加算
        loss_train+=loss.item()
        #勾配を0に初期化
        optimizer.zero_grad()
        #逆伝播
        loss.backward()
        #パラメータの更新
        optimizer.step()
    #1バッチあたりの平均の誤差
    loss_train=loss_train/(j+1)
    #record_loss_trainにappend
    record_loss_train.append(loss_train)
    
    ###検証モード###
    #順伝播による予測
    y_valid=net(x_valid)
    #誤差の算出
    loss_valid=loss_function(y_valid, t_valid)
    #record_loss_validにappend
    record_loss_valid.append(loss_valid)
    
    if i%1==0:
        print(f"Epoch {i}: Loss_train={loss_train}, Loss_valid={loss_valid}")
```
